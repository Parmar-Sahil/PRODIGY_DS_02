```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
Titanic = pd.read_csv('Task-2/Titanic-Dataset.csv')
Titanic.head()
```

```python
Titanic.isnull().sum()
```

```python
Titanic.shape
```

```python
Titanic.describe()
```

```python
Titanic['Age'].fillna(Titanic['Age'].mean())
```

```python
sns.histplot(Titanic['Age'], kde=True)
plt.title('Age Count')
plt.xlabel('Age')
plt.ylabel('total')
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.countplot(data=Titanic, x="Sex", hue="Survived", palette="mako")
plt.title("Gender Survival")
plt.xlabel("Gender")
plt.ylabel(" Survived Count")
plt.legend(title="Survived: Survived=1 , Not_Survived=0")
plt.show()
```

```python

plt.figure(figsize=(6, 3))
sns.scatterplot(data=Titanic, x="Age", y="Fare", hue="Survived")
plt.title("Scatter Plot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived:")
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.barplot(data=Titanic, x="Pclass", y="Age", hue="Survived")
plt.title("Passenger Class vs Age")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.legend(title="Survived People")
```
